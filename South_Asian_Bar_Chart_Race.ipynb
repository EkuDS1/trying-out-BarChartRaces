## Import Libraries


```python
import pandas as pd
import bar_chart_race as bcr
import os
```

## Load Data

*Dataset courtesy of The World Bank*  
***Dataset name:*** *Literacy rate, adult total (% of people ages 15 and above)*  
***Source:*** *UNESCO Institute for Statistics ( uis.unesco.org ). Data as of September 2021*  
***Downloaded from*** *https://data.worldbank.org/indicator/SE.ADT.LITR.ZS*

The original names of the files were too long so I shortened them. I deleted some of the rows at the beginning of the literacy rates csv file which contained some metadata.

```python
# Load Data
df = pd.read_csv('wb_literacy_rates.csv')
dfMeta = pd.read_csv('country_metadata.csv')
```

## Preprocess Data

```python
# Get all aggregate categories like total literacy rate of Asia or Europe
aggrCategories = dfMeta[dfMeta['Region'].isna()]['Country Code']
# Drop unneeded rows containing aggregate categories
df = df.drop(df[df['Country Code'].isin(aggrCategories)].index)

# Drop unneeded columns since we only need Country Name and Year variables
df = df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)

# Set index to Country Name
df = df.set_index('Country Name')

# Transpose dataframe to get wide data format
df = df.T

# Remove 'Country Code' from column name and change index name to Year
df.columns.names = ['']
df.index.names = ['Year']

# Select some South Asian countries for comparison
df = df[['Afghanistan', 'Pakistan', 'India', 'Bangladesh', 'Sri Lanka']]
# Interpolate where there are missing values
df = df.interpolate()

# Drop columns where all values are NaN and replace NaNs with 0.0
df = df.dropna(how='all')
df = df.fillna(0.0)
```

```python
bcr.bar_chart_race(df=df, 
                   title='South Asian Country Literacy Rates (World Bank)')
```
